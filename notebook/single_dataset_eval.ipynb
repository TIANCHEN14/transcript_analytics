```python
import json
import matplotlib.pyplot as plt
import pandas as pd
from tkinter import Tk
from tkinter.filedialog import askopenfilename, askdirectory
import os
from pathlib import Path

Tk().withdraw()
```

```python
# create main list to store all the information
main_list = []
```

```python
# file directory of all the file
file_dir = askdirectory()


for root, _ , files in os.walk(file_dir):
    for file in files:
        # join the file path
        file_path = os.path.join(file_dir , file)
        
        # refresh templist
        temp_list = []
        
        # load json file
        with open (file_path , "r") as f:
            temp_list = json.load(f)
            
    
        main_list = main_list + temp_list
        
        
print(len(main_list))
```

```python
main_df = pd.DataFrame()

for index, content in enumerate(main_list):
    content_df = pd.DataFrame([content])
    
    main_df = pd.concat([main_df , content_df])
    
print(main_df.columns)
```

```python
main_df['clasiifer'].value_counts()
```

```python
counts = main_df['clasiifer'].value_counts()
print(type(counts))
```

```python
plt.pie(counts, labels = counts.index , autopct='%1.1f%%' , pctdistance=1.7, labeldistance=2)
plt.title("Percentage for classifer")
plt.show()
```
